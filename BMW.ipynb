```python
import pandas as pd # pandas
import matplotlib.pyplot as plt #matplot graficos
import seaborn as sns # seaborn para graficos
# df=dataframe
df= pd.read_csv('BMW.csv', sep = ';', encoding='latin1')
```

```python
df.info()
```

```python
df
```

```python
model_year_table = pd.crosstab(df['MODELO'], df['ANO'])
print("Tabela de Modelos por Ano:")
print(model_year_table)

plt.figure(figsize=(15, 8))
sns.heatmap(model_year_table, cmap="YlGnBu", annot=True, fmt="d", linewidths=.5)
plt.title('DistribuiÃ§Ã£o de Modelos por Ano')
plt.xlabel('Ano')
plt.ylabel('Modelo')
plt.show()
```

```python
model_region_table = pd.crosstab(df['MODELO'], df['REGIÃO'])
print("Tabela de Modelos por RegiÃ£o:")
print(model_region_table)

plt.figure(figsize=(15, 8))
sns.heatmap(model_region_table, cmap="Greens", annot=True, fmt="d", linewidths=.5)
plt.title('DistribuiÃ§Ã£o de Modelos por RegiÃ£o')
plt.xlabel('RegiÃ£o')
plt.ylabel('Modelo')
plt.show()
```

```python
year_fuel_table = pd.crosstab(df['ANO'], df['TIPO COMBUSTIVEL'])
print("Tabela de Anos por Tipo de CombustÃ­vel:")
print(year_fuel_table)

plt.figure(figsize=(15, 8))
year_fuel_table.plot(kind='bar', stacked=True, figsize=(15, 8))
plt.title('DistribuiÃ§Ã£o de Tipos de CombustÃ­vel por Ano')
plt.xlabel('Ano')
plt.ylabel('NÃºmero de VeÃ­culos')
plt.xticks(rotation=45, ha='right')
plt.legend(title='CombustÃ­vel')
plt.tight_layout()
plt.show()
```

```python
year_km_table = df.groupby('ANO')['Quilometragem_KM'].mean().reset_index()
print("Tabela de MÃ©dia de Quilometragem por Ano:")
print(year_km_table)

plt.figure(figsize=(15, 8))
sns.lineplot(x='ANO', y='Quilometragem_KM', data=year_km_table, marker='o')
plt.title('MÃ©dia de Quilometragem por Ano')
plt.xlabel('Ano')
plt.ylabel('MÃ©dia de Quilometragem (KM)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
fuel_sales_table = df.groupby('TIPO COMBUSTIVEL')['Volume_Vendas'].sum().reset_index()
print("Tabela de Volume de Vendas por Tipo de CombustÃ­vel:")
print(fuel_sales_table)

plt.figure(figsize=(10, 6))
sns.barplot(x='TIPO COMBUSTIVEL', y='Volume_Vendas', data=fuel_sales_table, palette='viridis')
plt.title('Volume de Vendas por Tipo de CombustÃ­vel')
plt.xlabel('Tipo de CombustÃ­vel')
plt.ylabel('Volume de Vendas')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
